```python
url = 'https://s3.eu-west-1.amazonaws.com/data.cyber.org.il/virtual_courses/introdata/colab/youth_survey_preprocessed.csv'
csv_notebook = pd.read_csv(url)
```

```python
count_5 = ((csv_notebook['Opera'] == 5.0) & (csv_notebook['Age'] >= 18)).sum()
print(count_5)
```

```python
opera_5 = csv_notebook[(csv_notebook['Opera'] == 5.0) & (csv_notebook['Age'] >= 18)]
opera_5_and_gender=opera_5[['Gender']]
female_count=(opera_5_and_gender['Gender']=='female').sum()
male_count=(opera_5_and_gender['Gender']=='male').sum()
print(female_count)
print(male_count)
```
